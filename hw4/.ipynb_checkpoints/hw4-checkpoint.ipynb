```python
%matplotlib inline

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma, poisson,lognorm
from tqdm import trange
```

```python
data = []

datContent = [i.strip().split(',') for i in open("./premier_league_2013_2014.dat").readlines()]

for i in range(len(datContent)):
    data.append(list(map(float, datContent[i])))
    
data = np.array(data).astype(int)
```

```python
def log_norm(x,rsigma):
    return -0.5*x*np.sum(rsigma**2)

def log_gamma(x,alpha,beta):
    return (alpha-1)*np.log(x) - (1/beta)*x

def log_poisson(x,lam):
    return x*lam - np.exp(lam)
```

```python
fig, axes = plt.subplots(3,4,figsize=(20,15))
fig_idx = 0

for sigma in [0.005,0.05,0.5]:
    for t_interval in[1,5,20,50]: 
        T = 5000 + 5000*t_interval

        tau0 = 1e-4
        tau1 = 1e-4
        alpha = 0.1
        beta = 0.1


        burn_in_home = []
        sample_home = []

        posterior_dist_MH = [np.random.uniform(0, 1,size=45)]
        for t in trange(T):
            current = posterior_dist_MH[-1]
            proposal = np.zeros(45)
            for i in range(45):
                proposal[i] = np.random.normal(loc=current[i], scale=sigma)

            current_prob = 0
            proposal_prob = 0

            home_current = current[0]
            home_proposal = proposal[0]

            atts_current = current[1:21]
            atts_proposal = proposal[1:21]

            defs_current = current[21:41]
            defs_proposal = proposal[21:41]

            mu_att_current = current[41]
            mu_att_proposal = proposal[41]

            mu_def_current = current[42]
            mu_def_proposal = proposal[42]

            tau_att_current = current[43]
            tau_att_proposal = proposal[43]

            tau_def_current = current[44]
            tau_def_proposal = proposal[44]



            # home


            current_prob = current_prob + log_norm(home_current, tau0)
            proposal_prob = proposal_prob + log_norm(home_proposal, tau0)


            #atts, defs
            for n in range(20):
                current_prob = current_prob + log_norm(atts_current[n] - mu_att_current, tau_att_current)
                current_prob = current_prob + log_norm(defs_current[n] - mu_def_current, tau_def_current)
                proposal_prob = proposal_prob + log_norm(atts_proposal[n] - mu_att_proposal, tau_att_proposal)
                proposal_prob = proposal_prob + log_norm(defs_proposal[n] - mu_def_proposal, tau_def_proposal)


            # mu
            current_prob = current_prob + log_norm(mu_att_current, tau1)
            current_prob = current_prob + log_norm(mu_def_current, tau1)
            proposal_prob = proposal_prob + log_norm(mu_att_proposal, tau1)
            proposal_prob = proposal_prob + log_norm(mu_def_proposal, tau1)


            #tau
            current_prob = current_prob + log_gamma(tau_att_current, alpha, beta)
            current_prob = current_prob + log_gamma(tau_def_current, alpha, beta)
            proposal_prob = proposal_prob + log_gamma(tau_att_proposal, alpha, beta)
            proposal_prob = proposal_prob + log_gamma(tau_def_proposal, alpha, beta)


            #y    
            for g in range(380):
                current_prob = current_prob + log_poisson(data[g][0], home_current + atts_current[data[g][2]] - defs_current[data[g][3]])
                current_prob = current_prob + log_poisson(data[g][1], atts_current[data[g][3]] - defs_current[data[g][2]])
                proposal_prob = proposal_prob + log_poisson(data[g][0], home_proposal + atts_proposal[data[g][2]] - defs_proposal[data[g][3]])
                proposal_prob = proposal_prob + log_poisson(data[g][1], atts_proposal[data[g][3]] - defs_proposal[data[g][2]])



            accept_prob = min(np.exp(proposal_prob - current_prob), 1)

            if np.random.uniform(0,1) < accept_prob:
                posterior_dist_MH.append(proposal)
            else:
                posterior_dist_MH.append(current)

            if t<5000:
                burn_in_home.append(posterior_dist_MH[-1][0])
            elif t%t_interval==0:
                sample_home.append(posterior_dist_MH[-1][0])
                
        axes[fig_idx // 4, fig_idx % 4].plot(burn_in_home,label='Burn-in')
        axes[fig_idx // 4, fig_idx % 4].plot(sample_home,label='Sample')
        axes[fig_idx // 4, fig_idx % 4].legend()
        axes[fig_idx // 4, fig_idx % 4].set(title='Trace, sigma='+ str(sigma) + ', t=' +str(t_interval))


        fig_idx = fig_idx + 1
            
        
```
