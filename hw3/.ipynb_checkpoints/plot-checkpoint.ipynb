```python
import matplotlib.pyplot as plt
import numpy as np
train_l = [38.42778907775879, 34.44132444190979, 33.880603281497955]
test_l = [35.25559211881451, 34.282492273667664, 33.89866800332764]
train_s = [1.8219443359375, 1.59384716796875, 1.5415809592088336]
test_s = [1.8644923215051459, 1.6731078473413379, 1.55786865234375]
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot(range(1,4),train_l,label='Training Loss')
plt.plot(range(1,4),test_l,label='Testing Loss')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('FCN + l-SVM')

k = 1
plt.subplot(1,2,2)
plt.plot(range(1,4),k*np.array(train_s),label='Training Loss')
plt.plot(range(1,4),k*np.array(test_s),label='Testing Loss')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('FCN + s-SVM')


```
